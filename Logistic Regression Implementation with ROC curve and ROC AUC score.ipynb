Logistic Regression for Imbalanced Datset

```python
#Generate and plot a synthetic imbalnced classification dataset
from collections import Counter
from sklearn.datasets import make_classification
```

```python
#Imbalanced Datasets
x,y = make_classification (n_samples = 1000, n_features = 2, n_clusters_per_class = 1, n_redundant = 0,weights=[0.99],random_state=10)
```

```python
x
```

```python
Counter(y)
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame(x, columns=['feature_1', 'feature_2'])
sns.scatterplot(x='feature_1', y='feature_2', hue=y, data=df)
```

```python
#Split the datatsets into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
#Hyperparameter Tuning
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
penalty=['11','12','elasticNet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
class_weight = [{0:w,1:y}for w in [1,10,50,100]for y in[1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression

# Define the logistic regression model
model = LogisticRegression()

# Define stratified k-fold cross-validation
cv = StratifiedKFold(n_splits=5)

# Define the parameter grid
params = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'C': [100, 10, 1.0, 0.1, 0.01],
    'solver': ['liblinear', 'saga']  # Note: 'l1' and 'elasticnet' penalties require these solvers
}

# Initialize GridSearchCV
grid = GridSearchCV(estimator=model, 
                    param_grid=params, 
                    scoring='accuracy', 
                    cv=cv)

# Fit the grid search to the training data
grid.fit(x_train, y_train)

# Print the best parameters and score
print("Best parameters found: ", grid.best_params_)
print("Best score found: ", grid.best_score_)
```

```python

grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = grid.best_estimator_.predict(x_test)
score = accuracy_score(y_test, y_pred)
print("Accuracy score: ", score)
print("Classification report:\n", classification_report(y_test, y_pred, zero_division=0))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
```
